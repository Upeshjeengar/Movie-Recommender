###LowerCasing

```python
import pandas as pd
df=pd.read_csv('movie_data.csv')
df['overview'] = df['overview'].str.lower()
df['genres']=df['genres'].astype(str)
df['genres'] = df['genres'].str.lower()
df.head()
```

###Remove Punctuations

```python
import string
exclude=string.punctuation
df1=df
def remove_punc(text):
  if isinstance(text, str):
    for char in exclude:
      text=text.replace(char,'')
  return text

for i in range(0,9233):
  df['overview'][i]=remove_punc(df['overview'][i])
  #we can observe [,],, and ' are only punctuation so directly replacing them
  df['genres'][i] = df['genres'][i].replace(',', '').replace("'", '').replace("[", '').replace("]", '')

df.head()
```

###Remove stop Words


```python
!pip install nltk -q
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopwords=stopwords.words('english')

def remove_stopwords(text):
    if isinstance(text, str):  # Check if text is a string
        new_text = []
        for word in text.split():
            if word not in stopwords:
                new_text.append(word)
        return ' '.join(new_text)
    elif isinstance(text, list):  # Check if text is a list
        new_text_list = []
        for sentence in text:
            new_sentence = []
            for word in sentence.split():
                if word not in stopwords:
                    new_sentence.append(word)
            new_text_list.append(' '.join(new_sentence))
        return new_text_list
    else:
        return text

for i in range(0,9233):
  df['overview'][i]=remove_stopwords(df['overview'][i])

df
```

###Remove emojis


```python
import re #regular expression
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

for i in range(0,9233):
  df['overview'][i]=remove_emoji(df['overview'][i])
df.head()
```

###Stemmer


```python
from nltk.stem.porter import PorterStemmer
df=df1
```

```python
df.head()
```

```python
ps = PorterStemmer()
def stem_words(text):
    if isinstance(text, str):  # Check if text is a string
        return " ".join([ps.stem(word) for word in text.split()])
    elif isinstance(text, list):  # Check if text is a list
        return [stem_words(item) for item in text]
    return text

for i in range(0,9233):
  df['genres'][i]=stem_words(df['genres'][i])
  df['overview'][i]=stem_words(df['overview'][i])
df.head()
```
